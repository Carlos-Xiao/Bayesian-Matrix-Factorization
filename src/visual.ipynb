```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('prototype_results_D=10.csv')

iterations = df['Iteration']
train_mae = df['Train_MAE']
test_mae = df['Test_MAE']
train_rmse = df['Train_RMSE']
test_rmse = df['Test_RMSE']

plt.plot(iterations, train_mae, label='Train_MAE')
plt.plot(iterations, test_mae, label='Test_MAE')
plt.plot(iterations, train_rmse, label='Train_RMSE')
plt.plot(iterations, test_rmse, label='Test_RMSE')

plt.xlabel('Iteration')
plt.ylabel('Metrics')
plt.legend()
plt.show()
```

```python
D_values = [2, 5, 10, 20, 50, 100]

fig, ax = plt.subplots(figsize=(10, 6))

for D in D_values:
    file_path = f'full_results_D={D}.csv'
    df = pd.read_csv(file_path)

    ax.plot(df['Iteration'], df['Test_MAE'], label=f'D={D} MAE')
    ax.plot(df['Iteration'], df['Test_RMSE'], label=f'D={D} RMSE', linestyle='--')

ax.set_xlabel('Iteration')
ax.set_ylabel('Metrics')
ax.legend()

plt.show()
```

```python
df = pd.read_csv('gmm_results_D=10.csv')
df_reshaped = pd.DataFrame()
df_reshaped['Iteration'] = df['Iteration']

for i in range(4):
    df_reshaped[f'Train_MAE_{i}'] = df[f'Train_MAE_{i}']
    df_reshaped[f'Test_MAE_{i}'] = df[f'Test_MAE_{i}']
    df_reshaped[f'Train_RMSE_{i}'] = df[f'Train_RMSE_{i}']
    df_reshaped[f'Test_RMSE_{i}'] = df[f'Test_RMSE_{i}']

df_reshaped['Tot_Train_MAE'] = df['Tot_Train_MAE']
df_reshaped['Tot_Test_MAE'] = df['Tot_Test_MAE']
df_reshaped['Tot_Train_RMSE'] = df['Tot_Train_RMSE']
df_reshaped['Tot_Test_RMSE'] = df['Tot_Test_RMSE']

plt.figure(figsize=(10, 6))

for i in range(4):
    plt.plot(df_reshaped['Iteration'], df_reshaped[f'Train_MAE_{i}'], label=f'Train_MAE_{i}')
    plt.plot(df_reshaped['Iteration'], df_reshaped[f'Test_MAE_{i}'], label=f'Test_MAE_{i}')
    plt.plot(df_reshaped['Iteration'], df_reshaped[f'Train_RMSE_{i}'], label=f'Train_RMSE_{i}')
    plt.plot(df_reshaped['Iteration'], df_reshaped[f'Test_RMSE_{i}'], label=f'Test_RMSE_{i}')

plt.plot(df_reshaped['Iteration'], df_reshaped['Tot_Train_MAE'], label='Tot_Train_MAE', linestyle='--')
plt.plot(df_reshaped['Iteration'], df_reshaped['Tot_Test_MAE'], label='Tot_Test_MAE', linestyle='--')
plt.plot(df_reshaped['Iteration'], df_reshaped['Tot_Train_RMSE'], label='Tot_Train_RMSE', linestyle='--')
plt.plot(df_reshaped['Iteration'], df_reshaped['Tot_Test_RMSE'], label='Tot_Test_RMSE', linestyle='--')

plt.xlabel('Iteration')
plt.ylabel('Metrics')
plt.legend()
plt.show()
```
